```python
from pymongo import MongoClient
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
db, coll = connectCollection('companies', 'companies')
```

```python
print(coll.json)
```

```python
comp = list(coll.find({'$and': [{"offices.latitude": { '$nin': [None] }}, {"offices.longitude":{ '$nin': [None] }}, {"offices.city": "Amsterdam"}, {'deadpooled_year': None}]}))
```

```python
def getLocation(companies, x):
    longitude = companies['offices'][x]['longitude']
    latitude = companies['offices'][x]['latitude']
    loc = {
        'type':'Point',
        'coordinates':[float(longitude), float(latitude)]
    }
    return loc
```
