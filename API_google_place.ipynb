```python
import googlemaps
from datetime import datetime
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
gmaps = googlemaps.Client(key=os.getenv("key"))

# Geocoding an address
places_result = gmaps.places('Starbucks, Amsterdam')
```

```python
def getLocation(companies):
    location =[]
    for i in range(len(companies['results'])):
        longitude = companies['results'][i]['geometry']['location']['lng']
        latitude = companies['results'][i]['geometry']['location']['lat']
        name=companies['results'][i]['name']
        loc = {
            'name': name,
            'type':'Point',
            'coordinates':[longitude, latitude]
        }
        location.append(loc)
    return location
```

```python
getLocation(places_result)
```

```python
school = gmaps.places('school, Amsterdam')
```

```python
print(len(school['results']))
getLocation(school)
```
