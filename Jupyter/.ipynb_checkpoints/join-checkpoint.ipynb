```python
map_city=folium.Map(location=[52.37021,4.895168], zoom_start=12)
for cmp in comp2:
    for i in range(len(cmp['location'])):
        folium.Marker(cmp['location'][i]['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='usd',color='red'), 
                   ).add_to(map_city)

for al in aloc:
    folium.Marker(al['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='plane',color='black'), 
                   ).add_to(map_city)

for kd in kdloc:
    folium.Marker(kd['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='heart',color='blue'), 
                   ).add_to(map_city)

for nc in ncloc:
    folium.Marker(nc['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='glass',color='purple'), 
                   ).add_to(map_city)

for sb in sbloc:
    folium.Marker(sb['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(prefix="fa", icon='coffee',color='darkgreen'), 
                   ).add_to(map_city)

for sc in scloc:
    folium.Marker(sc['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='book',color='orange'), 
                   ).add_to(map_city)

for vr in vrloc:
    folium.Marker(vr['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='cutlery',color='cadetblue'), 
                   ).add_to(map_city)

for x1 in x:
    folium.Marker(x1['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='home',color='beige'), 
                   ).add_to(map_city)

for sp in shoploc:
    folium.Marker(sp['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='leaf',color='green'), 
                   ).add_to(map_city)

    
# color_options = {'beige', 'black', 'blue', 'cadetblue', 'darkblue', 'darkgreen', 'darkpurple', 'darkred',
# 'gray', 'green', 'lightblue', 'lightgray', 'lightgreen', 'lightred', 'orange', 'pink', 'purple', 'red', 'white'}
```
